```python
#Q2 - How are class-based exceptions that have been raised matched to handlers?
#AnIn Python, class-based exceptions that have been raised are matched to handlers based on their inheritance hierarchy. When an exception is raised, Python starts searching for an appropriate exception handler by examining the except blocks in the code.

Here's how the matching process works:

- Python checks the except blocks in the order they appear in the code, from top to bottom.

- For each except block, Python compares the raised exception's class (or one of its base classes) to the specified exception type in the except statement. It checks if the raised exception is an instance of the specified exception type or a subclass thereof.

- If a matching exception type is found, the corresponding except block is executed, and the exception is considered handled. Python does not continue searching for matching handlers beyond the first match.

- If no matching exception type is found in any of the except blocks, the exception propagates to the next outer exception handler or terminates the program if no outer handler is found.

#Exp:

class CustomException(Exception):
    pass

class DerivedException(CustomException):
    pass

try:
    raise DerivedException("An exception occurred")
except CustomException:
    print("CustomException handler")
except Exception:
    print("Generic Exception handler")
```

```python
#Q3 - Describe two methods for attaching context information to exception artefacts.
#Answer:
Attaching context information to exception artifacts in Python can be useful for providing additional details about the exception, 
aiding in debugging, and providing meaningful error messages. Here are two common methods for attaching context information to exception 
artifacts:

1. Exception Arguments: 
#Exp:
class CustomException(Exception):
    def __init__(self, message, error_code):
        super().__init__(message)
        self.error_code = error_code

try:
    raise CustomException("Something went wrong", 500)
except CustomException as e:
    print(e)
    print("Error code:", e.error_code)


2. Exception Context Managers:
#Exp:
class ExceptionContextManager:
    def __enter__(self):
        pass

    def __exit__(self, exc_type, exc_val, exc_tb):
        if exc_val is not None:
            exc_val.context = "Additional context information"

try:
    with ExceptionContextManager():
        # Code that may raise an exception
        raise ValueError("Something went wrong")
except ValueError as e:
    print(e)
    print("Context:", e.context)
```
